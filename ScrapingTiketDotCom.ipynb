```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import time
import json
```

```python
# Save data
data_total = []
impression_total = []
nama_total = []
location_total = []
city_total = []
rating_total = []
price_total = []
description_total = []
img_total = []
gmaploc_total = []
lon_total = []
lat_total = []
gambar_total = []
desk_total = []

for i in range(0,100):
    #Url Scraping Jawa Barat
    if i == 0 :
        url = f'https://www.tiket.com/hotel/indonesia/region/west-java-108001534490276256'
    else :
        url = f'https://www.tiket.com/hotel/indonesia/region/west-java-108001534490276256/page-{i+1}'
    print(url)
    print("Didalamnya ada")
    
    # Set headers
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36'
    }

    # Get response using requests.get()
    r = requests.get(url, headers=headers)

    # Parse the text using BeautifulSoup
    soup = BeautifulSoup(r.text, 'html.parser')

    # Make items
    items = soup.find_all('div',{'class':'card-frame'})

    u = 0
    for it in items:
        #Get url Hotel / page
        # print(it)
        data = it.find('a').get('href')
        data = "https://www.tiket.com"+ data
        data_total.append(data)
    
        urls = data

        # Get response using requests.get()
        r = requests.get(urls, headers=headers)

        # Parse the text using BeautifulSoup
        soup = BeautifulSoup(r.text, 'html.parser')

        # Make item
        item = soup.find_all('div', {'class':'main_hotel_detail__ZyWLS'})
        u = u + 1
        for its in item:
            
            print(urls)
            print(f'Response: {r.status_code} for url hotel: {u}')
            if r.status_code != 200 :
                break
            # Get gmap Url
            try :
                gmaploc = its.find('a', {'class':'LocationSection_see_maps__PiVCo HcPVsG_text HcPVsG_variant_activity HcPVsG_size_b2 HcPVsG_weight_bold'}).get('href')
            except :
                gmaploc = np.nan
            gmaploc_total.append(gmaploc)

            if gmaploc is np.nan :
                # Get longitude
                lon = 0
                lon_total.append(lon)

                # Get latitude
                lat = 0
                lat_total.append(lat)
            else :
                lonlat = gmaploc[48:].split(",")
                # Get longitude
                lon = lonlat[0]
                lon_total.append(lon)

                # Get latitude 
                lat = lonlat[1]
                lat_total.append(lat)

            # Get image
            soup = BeautifulSoup(requests.get(urls).content,'html.parser')
            data = [json.loads(x.string) for x in soup.find_all('script', type= "application/ld+json")]
            for d in data:
                for obj in d:
                    if "image" in obj:
                        gambar =obj["image"]
                    if "mainImage" in obj:
                        gambar =obj["mainImage"]  
                    if gambar == "/" or gambar == "":
                        gambar = 0
                    if "description" in obj:
                        desk = obj["description"]
            gambar_total.append(gambar)
            desk_total.append(desk)
            
        # Get name
        nama = it.find('h3', {'class':'title ellipsis'}).text
        nama_total.append(nama)

        # Get Lokasi
        loc = it.find('div', {'class':'location ellipsis'}).text.split(',')[0]
        location_total.append(loc)

        # Get City
        try :
            city = it.find('div', {'class':'location ellipsis'}).text.split(',')[1]
        except :
            city = it.find('div', {'class':'location ellipsis'}).text.split(',')[0]
        if city[:1] == " ":
            city = city[1:]
        city_total.append(city)
        
        # Get rating
        try:
            rating = float(it.find('div', {'class':'tiket-rating'}).text)
        except:
            rating = 0
        rating_total.append(rating)

        # Get impression
        try:
            impression = it.find('div', {'class':'rating'}).find('div', {'class':'tiket-impression'}).text
        except:
            impression = 0
        impression_total.append(impression)
        
        # Get price
        try:
            price = int(''.join(it.find('div', {'class':'price-wrap'}).find('div', {'class':'after-price'}).text.split('IDR ')[-1].split('.')))
        except:
            price = 0
        price_total.append(price)
        
    print(f'Response: {r.status_code} for page: {i+1}')
    print(len(data_total))
    jumlah=[len(nama_total),len(location_total),len(lon_total),len(lat_total),len(gmaploc_total),len(rating_total),len(impression_total),len(price_total),len(gambar_total),len(data_total),len(desk_total)]
    print(jumlah)
```

```python
jumlah=[len(nama_total),len(location_total),len(lon_total),len(lat_total),len(gmaploc_total),len(rating_total),len(impression_total),len(price_total),len(gambar_total),len(data_total),len(desk_total)]
print(jumlah)
df = pd.DataFrame(data={'nama':nama_total,
                        'location':location_total,
                        'city':city_total,
                        'longitude':lon_total,
                        'latitude':lat_total,
                        'gmap':gmaploc_total,
                        'rating':rating_total,
                        'impression':impression_total,
                        'price':price_total,
                        'gambar':gambar_total,
                        'url':data_total,
                        'deskripsi':desk_total})
df.head()
res = df[~df.duplicated('nama')]
res = res.loc[~(df==0).any(axis=1)]
res.to_excel('hotel_jawabarat_tiket_fix.xlsx',index=False)
res.to_csv('hotel_jawabarat_tiket_fix1.csv',index=False)
df.to_excel('hotel_jawabarat_tiket_fix2.xlsx',index=False)
```
